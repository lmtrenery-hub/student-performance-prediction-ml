# Session 3: Python and Google Colab Setup
This notebook checks whether the standard Python data-science libraries import
correctly and records their versions for reproducibility.

```python
import sys
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
import sklearn

print("Python", sys.version.split()[0])

for m in (pd, np, matplotlib, sns, sklearn):
    print(m.__name__, m.__version__)
```

## Environment Check Result
The Python environment successfully imported pandas, numpy, matplotlib, seaborn,
and scikit-learn.
The printed versions document the computational environment used for this project.
