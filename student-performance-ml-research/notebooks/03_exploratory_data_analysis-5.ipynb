````python
support_df = df[["schoolsup", "famsup", "higher", "G3"]].copy()

# Standardize the text categories
for col in ["schoolsup", "famsup", "higher"]:
    support_df[col] = (
        support_df[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

# Ensure G3 is numeric
support_df["G3"] = pd.to_numeric(
    support_df["G3"],
    errors="coerce"
)

# Remove rows with missing required values
support_df = support_df.dropna(
    subset=["schoolsup", "famsup", "higher", "G3"]
)

print("Original number of rows:", len(df))
print("Rows used in the support analysis:", len(support_df))
display(support_df.head())
````

````python
for col in ["schoolsup", "famsup", "higher"]:
    mean_grades = support_df.groupby(col)["G3"].mean().round(2)

    print("=" * 30)
    print(f"Grouped by: {col}")
    print("-" * 30)

    # Using format alignment: < for left, > for right
    for index, value in mean_grades.items():
        print(f"{str(index):<15} {value:>10}")
    print()
````

````python
for col in ["schoolsup", "famsup", "higher"]:
    summary = (
        support_df
        .groupby(col)["G3"]
        .agg(
            student_count="count",
            mean_G3="mean",
            median_G3="median",
            standard_deviation="std"
        )
        .round(2)
        .reset_index()  # This converts the index into a regular column
    )

    print("=" * 60)
    print(f"Grade summary by {col}")
    display(summary)
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Rebuild the support_summary DataFrame
summary_tables = []
for col in ["schoolsup", "famsup", "higher"]:
    temp = (
        support_df
        .groupby(col)["G3"]
        .agg(student_count="count", mean_G3="mean", median_G3="median", standard_deviation="std")
        .reset_index()
    )
    temp = temp.rename(columns={col: "response"})
    temp.insert(0, "feature", col)
    summary_tables.append(temp)

support_summary = pd.concat(summary_tables, ignore_index=True)
support_summary[["mean_G3", "median_G3", "standard_deviation"]] = support_summary[["mean_G3", "median_G3", "standard_deviation"]].round(2)

# 2. Plotting the consolidated data
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

plot = sns.barplot(
    data=support_summary,
    x="feature",
    y="mean_G3",
    hue="response",
    palette="viridis"
)

plt.title("Comparison of Mean G3 Grades by Support Feature", fontsize=14)
plt.xlabel("Support Feature Type", fontsize=12)
plt.ylabel("Mean Final Grade (G3)", fontsize=12)
plt.legend(title="Response (Yes/No)")

plt.tight_layout()
plt.show()
````

As your EDA assistant, I have analyzed the relationship between your selected support features (`schoolsup`, `famsup`, `higher`) and the final grade (`G3`). Below is the evaluation for your feature engineering process.

---

### 1. School Support (`schoolsup`)

* **Summary:** Students receiving extra educational support from the school typically show lower mean final grades compared to those who do not.
* **Interpretation:** This is a classic example of **selection bias**. School support is generally assigned to students who are already struggling academically. The negative correlation does not imply the support is harmful, but rather that the support is successfully being targeted at the students who need it most.
* **Recommendation:** **Keep.** However, do not treat this as a direct performance driver. In a machine learning model, this feature captures "at-risk" status.

### 2. Family Support (`famsup`)

* **Summary:** Students with family educational support often show slightly higher mean grades, though the gap is typically narrower than other factors.
* **Interpretation:** This feature represents social capital and a supportive home environment. It is a more direct indicator of a stable learning foundation outside of the classroom.
* **Recommendation:** **Keep.** It is a useful behavioral feature that likely adds predictive power regarding student attendance and engagement.

### 3. Aspiration for Higher Education (`higher`)

* **Summary:** This feature shows the most significant positive impact on final grades. Students who express a desire for higher education consistently outperform those who do not by a wide margin.
* **Interpretation:** This is an **attitudinal feature**. It acts as a strong proxy for student motivation and long-term goal setting, which are often better predictors of academic success than external support systems.
* **Recommendation:** **Keep (High Priority).** This is likely one of the most statistically significant predictors in your dataset.

The most "surprising" direction is the negative correlation for `schoolsup`. Beginners often mistakenly assume this feature "causes" lower grades. It is vital to document in your EDA report that this is **reverse causality**: the students were not struggling *because* they had support; they had support *because* they were struggling.

### Interpretation of Results

*
**Data Leakage Concern:** G1 and G2 are prior academic grades and are typically very strongly correlated with G3. Including them in a model designed for *early intervention* can create "temporal leakage," where the model "peeks" at future outcomes by using grades from the same academic period.


*
**The Two-Scenario Approach:** To handle this, your research should distinguish between:


1.
**Full-Information Model (Include G1, G2):** Uses all available data to measure maximum predictive performance.


2.
**Early-Warning Model (Exclude G1, G2):** Relies only on demographic/behavioral variables to identify at-risk students *before* later grades are known.





By comparing these two scenarios, you maintain scientific transparency, acknowledging that while including prior grades may improve statistical accuracy, excluding them is often necessary to provide enough time for meaningful instructor intervention.
