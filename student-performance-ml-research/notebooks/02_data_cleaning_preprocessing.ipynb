```python
# Session 11: Categorical Variables Analysis
import pandas as pd

# 1. Load the dataset directly from the root folder
df = pd.read_csv("student-mat.csv", sep=";")

# Confirm dataset is loaded
print("Dataset shape:", df.shape)

# Select categorical columns
cat_cols = df.select_dtypes(include="object").columns.tolist()
print("\nNumber of categorical columns:", len(cat_cols))
print("Categorical columns:", cat_cols)

print("\n" + "="*80)
print("Categorical column unique-value preview:")
for c in cat_cols:
    unique_values = df[c].dropna().unique()
    print("\n" + "-"*40)
    print("Column:", c)
    print("Number of unique values:", df[c].nunique(dropna=False))
    print("First 10 unique values:", unique_values[:10])

# Create summary table
cat_summary = []
for c in cat_cols:
    cat_summary.append({
        "column": c,
        "unique_count": df[c].nunique(dropna=False),
        "example_values": list(df[c].dropna().unique()[:10])
    })
cat_summary_df = pd.DataFrame(cat_summary)
cat_summary_df
```
