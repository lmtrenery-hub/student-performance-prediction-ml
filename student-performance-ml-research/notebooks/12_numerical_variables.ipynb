```python
# Identify numeric columns
num_cols = df.select_dtypes(include="number").columns.tolist()
print("Number of numeric columns:", len(num_cols))
print("Numeric columns:")
print(num_cols)
```

```python
# Preview numeric columns and their ranges
numeric_summary = df[num_cols].agg(["min", "max", "mean"]).T
print(numeric_summary)
```

```python
# Clean numeric summary table
numeric_summary = df[num_cols].agg(["min", "max", "mean"]).T
numeric_summary = numeric_summary.round(2)
display(numeric_summary)
```

```python
# Save numeric summary to reports folder
import os
os.makedirs("reports", exist_ok=True)
numeric_summary.to_csv("reports/session12_numeric_summary.csv")
print("Saved: reports/session12_numeric_summary.csv")
```
