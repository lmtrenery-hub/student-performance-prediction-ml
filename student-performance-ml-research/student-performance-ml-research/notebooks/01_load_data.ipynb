```python
import pandas as pd
# UCI student-mat.csv uses ; as the separator
df = pd.read_csv("student-mat.csv", sep=";")
print("Shape:", df.shape)
print(df.head())
```

### Summary

A CSV file has loaded correctly if the DataFrame contains the expected number of rows and columns and the data is displayed in separate columns. Two useful checks are `df.shape`, which confirms the dataset dimensions, and `df.head()`, which displays the first five rows so you can verify that the column names and values appear correctly.

### Interpretation

The first check is to confirm the DataFrame shape matches the expected dimensions of the dataset. The second check is to inspect the first few rows using `df.head()` to ensure the data is organized into the correct columns. If the wrong separator is used when loading the CSV, pandas may place an entire row into a single column, making the dataset unsuitable for analysis and machine learning.

### Recommendation

After confirming the file loaded correctly, inspect the dataset further by checking the column names, data types, and missing values using `df.columns`, `df.info()`, and `df.isnull().sum()`. This ensures the data is ready for cleaning, exploration, and model development.
