# Session 31: Regression Model Comparison

This notebook loads the regression leaderboard generated from the Session 25-30 model families, ranks models by RMSE, identifies the winner, and recreates the comparison figure.

Dataset used by the automation: `data/student-mat.csv`

`G1` and `G2` were excluded from the model predictors to prevent final-grade leakage.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
LEADERBOARD_PATH = PROJECT_ROOT / 'reports' / 'regression_leaderboard.csv'
FIGURE_PATH = PROJECT_ROOT / 'figures' / 'regression_rmse_comparison.png'

leaderboard = pd.read_csv(LEADERBOARD_PATH)
leaderboard = leaderboard.sort_values(['RMSE', 'MAE']).reset_index(drop=True)
leaderboard['Rank'] = range(1, len(leaderboard) + 1)
leaderboard = leaderboard[['Model', 'Rank', 'RMSE', 'MAE', 'R2']]
leaderboard
```

```python
winner = leaderboard.iloc[0]
print('Regression winner')
print('-----------------')
print(f"Model: {winner['Model']}")
print(f"RMSE: {winner['RMSE']:.4f}")
print(f"MAE:  {winner['MAE']:.4f}")
print(f"R2:   {winner['R2']:.4f}")
```

```python
top_three = leaderboard.head(3).copy()
best_rmse = top_three.iloc[0]['RMSE']
top_three['RMSE Gap from Best'] = top_three['RMSE'] - best_rmse
top_three['RMSE Gap Percentage'] = top_three['RMSE Gap from Best'] / best_rmse * 100
top_three
```

```python
chart_data = leaderboard.sort_values('RMSE', ascending=False)
winner_name = leaderboard.iloc[0]['Model']
colors = ['#2ca02c' if name == winner_name else '#1f77b4' for name in chart_data['Model']]

fig, ax = plt.subplots(figsize=(12, max(7, len(chart_data) * 0.58)))
bars = ax.barh(chart_data['Model'], chart_data['RMSE'], color=colors)
ax.set_title('Regression Model Comparison by RMSE')
ax.set_xlabel('Root Mean Squared Error - Lower Is Better')
ax.set_ylabel('Regression Model')
ax.grid(axis='x', alpha=0.25)

for bar, value in zip(bars, chart_data['RMSE']):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f' {value:.4f}', va='center')

fig.tight_layout()
fig.savefig(FIGURE_PATH, dpi=300, bbox_inches='tight')
plt.show()
```

## Interpretation

The model in rank 1 has the lowest test-set RMSE and is the current regression winner. The top-three RMSE gaps should also be examined because a very small difference may not represent a practically important advantage.
